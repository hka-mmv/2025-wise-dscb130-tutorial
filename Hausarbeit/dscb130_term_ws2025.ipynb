# Motivation
In den kommenden Semestern werden komplexe Berechnungen und Datenanalysen in nahezu allen Fachgebieten immer wichtiger werden. Grundbaustein dieser Analysen sind fundierte mathematische Kenntnisse und Programmierfertigkeiten. In diesem Projekt entwickeln die Studierenden in kleinen Teams einen vielseitigen grafischen (GUI) Taschenrechner in Python, mit dem Sie zentrale Aspekte der Mathematik - von einfachen Grundrechenarten bis hin zu komplexen Operationen - selbstständig implementieren.  
Das Projekt ist in mehrere Aufgabenblöcke unterteilt, für deren erfolgreiche Lösung Sie jeweils Klausurbonus-Punkte erhalten. Zum Ende des Semesters werden die erarbeiteten Inhalte präsentiert.  

Punkte gibt es **nur,** wenn eine Präsentation stattfindet und mindestens vier Punkte in der inhaltlichen Bearbeitung erreicht werden. So kann man also Aufgabe 1 bearbeiten und präsentieren, erhält dafür 6 Punkte *(4 durch Aufgabe 1, 2 durch die Präsentation)*, welche auf die Klausur angerechnet werden.

# Aufgabe 1 - Grundrechenarten und GUI-Struktur
1. GUI-Basics in tkinter:  
Erstellen eines Hauptfensters mit Python und tkinter (z. B. Tk()-Objekt erstellen). Layout definieren: z. B. Eingabezeile (Entry-Widget) oben, Tastenfeld mit Buttons darunter.
2. Implementierung der Grundrechenarten:  
Addition, Subtraktion, Multiplikation, Division über entsprechende Buttons. Programmierung erster Funktionen zur Berechnung (z. B. „+“, „-“, „×“, „÷“).
3. Interaktion mit dem Eingabefeld:  
Eingabe von Zahlen (z. B. durch direkte Tastatureingabe oder Klick auf die entsprechenden Ziffern-Buttons). Anzeige der eingegebenen Zahlen bzw. Rechenergebnisse im GUI.

## Erwartete Ergebnisse
- Ein funktionsfähiges Hauptfenster mit eingebauten Buttons für 0–9 und die Grundrechenarten
- Berechnungen über die GUI sind fehlerfrei möglich
- Strukturierter, kommentierter Code als Ausgangspunkt für die nächsten Aufgaben

## Hinweise und Ressourcen
- [Python Tkinter Dokumentation](https://docs.python.org/3/library/tkinter.html) für detaillierte Informationen zu Widgets und Layout-Methoden

# Aufgabe 2 - Mathematische Regeln
1. Implementierung der Punkt-vor-Strich-Logik:
Regelgerechtes Behandeln von Punkt- und Strichoperationen, nicht nur sequentielles Abarbeiten von mehreren Rechenoperatoren.
2. Erweiterung des Interfaces:
Ergänzung des Interfaces um eine öffnende und schließende Klammer.
3. Logik für Klammerbehandlung:
Das Vorhandensein von Klammern soll überprüft und priosisiert behandelt werden.

## Erwartete Ergebnisse
- Der Taschenrechner kann Ausdrücke mit mehreren Operatoren korrekt nach den mathematischen Prioritätsregeln (Punkt-vor-Strich) berechnen.
- Ausdrücke mit Klammern werden korrekt evaluiert.
- Die GUI wurde um die Tasten "(" und ")" erweitert.

# Aufgabe 3 - Verwenden von Konstanten und selbstdefinierebaren Variablen

1. Erweiterung um Konstanten:  
Konstanten sollen persistent sein und über eine externe JSON Datei in den Taschenrechner ladbar sein. 
Eine Vorgegebene JSON Konstanten Datei ist im GitHub Repository zu finden (constants.json).
2. Erweiterung des Eingabefeldes um eine Konstantenfunktion:  
Konstanten müssen über das grafische Interface abrufbar/einsehbar sein.
3. Erweiterung der Grundrechenarten mit selbstdefinierbaren Variablen:  
Möglichkeit der Variablenspeicherung und des Variablenaufrufs. Speicher muss nicht persistent sein (beim schließen des Taschenrechners können Variablen gelöscht werden). Variablen müssen in Rechenoperationen verwendbar sein.
4. Erweiterung des Eingabefeldes um eine Variablenfunktion:  
Variablen müssen über das grafische Interface abrufbar/einsehbar sein.

# Aufgabe 4 - Erweiterte Rechenfunktionen
1. Zusätzliche Buttons oder Menüs für die mathematischen Funktionen: 
Sinnvolle Layoutanpassungen (z.B. ein separates Panel für erweiterte Funktionen). Sicherstellung einer übersichtlichen Ergebnissdarstellung (z.B. Rundung)
- Trigonometrische Funktionen (z.B. sin, arccos)
- Exponential- und Logarithmusfunktionen z.B.(e, ln(x), log10(x))
- Wurzeln (sqrt, und beliebige Wurzeln)
- PI (begrenzt auf 3.14159) 

2. Implementierung von math-Funktionen mit korrekten Ausgaben:
Der Aufruf der Buttons soll auf eine funktionale Rechenlogik referenzieren und korrekte Ergebnisse liefern können.

## Hinweise und Ressourcen
- [Python Math](https://docs.python.org/3/library/math.html) für trigonometrische Berechnungen

## Erwartete Ergebnisse
- Funktionsfähiges erweitertes Rechnen

# Aufgabe 5 - Fehlerbehandlung
1. Erkennen von ungültigen Eingaben:
Fehlerhafte Eingaben (z.B. sqrt(-1) im Reellen Kontext, ln(0) oder Division durch 0), in Form von Klartextmeldungen (z.B. Pop-Up Fenster: „Eingabefehler: Logarithmus ist nicht definiert bei x ≤ 0“) darstellen.
2. Sperren von nicht erwünschten Funktionalitäten (bspw. Buttons ausgrauen, wenn ein Fehler angezeigt wird):
Umschalten des Taschenrechners in einen "inaktiven" Modus bei einer Fehlermeldung, um einfache Weiterausgaben zu verhindern. Gerne auch kreativ werden :)

## Hinweise und Ressourcen
- [Python Error Handling](https://blog.miguelgrinberg.com/post/the-ultimate-guide-to-error-handling-in-python)

## Erwartete Ergebnisse
- Das Programm stürzt nicht ab bei ungültigen mathematischen Operationen (insbesondere Division durch Null).
- Die Fehlermeldungen sind klar und dem Fehler angemessen
- Nach dem Auftreten eines Fehlers sind die Eingabe-Buttons gesperrt, bis der Benutzer den Reset-Button drückt.
