In dieser Aufgabe arbeiten Sie mit verschachtelten Datenstrukturen in Python. Schreiben Sie jeweils funktionierenden Python-Code, um die gestellten Anforderungen zu erfüllen. Es sollen Informationen aus einer Liste verarbeitet, aggregiert und gezielt verändert werden.

a) Dictionary aus einer Liste von Tupeln erstellen (2) Gegeben ist eine Liste von Bestellungen. Jeder Eintrag enthält den Namen eines Kunden, ein Getränk und den dazugehörigen Preis: 

```python
bestellungen = [ ("Anna", "Kaffee", 2.50),
                ("Anna", "Tee", 2.00),
                ("Bernd", "Wasser", 1.50),
                ("Bernd", "Kaffee", 2.50),
                ("Clara", "Tee", 2.00),
                ("Clara", "Kaffee", 2.50),
                ("Clara", "Saft", 3.00) ]
```

Erstellen Sie ein Dictionary namens kunden, in dem jeder Kundenname genau einmal vorkommt und jedem Kunden eine Liste seiner Bestellungen zugeordnet ist. Die Bestellungen sollen jeweils als Dictionary mit "getränk" und "preis" gespeichert werden. 
Beispiel für einen Eintrag: 

`"Anna": [ {"getränk": "Kaffee", "preis": 2.50}, {"getränk": "Tee", "preis": 2.00} ]`

```python
kunden = {}

for name, getraenk, preis in bestellungen:
    if name not in kunden:
        kunden[name] = []
    kunden[name].append({"getränk": getraenk, "preis": preis})

print(kunden)
```

## Unteraufgabe aus DSCB130 WiSe 2024 - Klausur

Unterscheiden Sie die sort-Funktion von der sorted-Funktion in Python anhand eines Programmbeispiels.

```python
zahlen = [5, 2, 3, 1, 4]

# Verwenden der sort-Funktion
zahlen.sort()
print(zahlen)  # Ausgabe: [1, 2, 3, 4, 5]

# Verwenden der sorted-Funktion
zahlen2 = [5, 2, 3, 1, 4]
sortierte_zahlen = sorted(zahlen2)
print(sortierte_zahlen)  # Ausgabe: [1, 2, 3, 4, 5]
print(zahlen2)  # Original bleibt unverändert: [5, 2, 3, 1, 4]
```

Geben Sie den Unterschied nun in Ihren eigenen Worten wieder.

A: Die sort-Funktion sortiert die Liste in-place und gibt None zurück, während die sorted-Funktion eine neue sortierte Liste zurückgibt und die ursprüngliche Liste unverändert lässt.
