## Lösung zu Aufgabe 1

```python
# Liste der Produkte mit Preis pro Kilo und Gewicht
produkte = [("Banane", 4.29, 1.0), ("Apfel", 4.95, 5.0), ("Mango", 14.99, 0.5)]

# Erstellung einer neuen Liste mit Gesamtpreisen
produkte_mit_preis = []

# Berechnung des Gesamtpreises und Hinzufügen zur neuen Liste
for name, preis, gewicht in produkte:
    gesamtpreis = preis * gewicht
    produkte_mit_preis.append((name, preis, gewicht, gesamtpreis))

# Sortieren nach Gesamtpreis (Index 3 in der Liste)
produkte_mit_preis.sort(key=lambda x: x[3]) # effektiv der einzige Lösungsweg ohne weitere Hilfsmittel

# Ausgabe der sortierten Liste mit Gesamtpreisen
for name, preis, gewicht, gesamtpreis in produkte_mit_preis:
    print(f"{name} (Preis pro Kilo: {preis}€, Gewicht: {gewicht}kg, Gesamtpreis: {gesamtpreis:}€)")
```

```python
# Musterlösung!!

produkte_mit_preis = [(name, preis_pro_kilo, gewicht, preis_pro_kilo * gewicht) for name, preis_pro_kilo, gewicht in produkte]

produkte_mit_preis.sort(key=lambda x: x[3])

for index, (name, preis_pro_kilo, gewicht, gesamtpreis) in enumerate(produkte_mit_preis, start=1):
    print(f"{index}. {name} - Preis pro Kilo: {preis_pro_kilo}€, Gewicht: {gewicht}kg, Gesamtpreis: {gesamtpreis}€")
```

```python
# Lösung aus dem Tutorium

# Hilfestellung; Liste der Produkte mit Preis pro Kilo und Gewicht
produkte = [("Banane", 4.29, 1.0), ("Apfel", 4.95, 5.0), ("Mango", 14.99, 0.5)]

# Gesamtpreis berechnen und in neuer Liste speichern
produkte_mit_gesamtpreis = []

for name, preis, gewicht in produkte: # iteriert über jedes objekt in jedem eintrag 
    gesamtpreis = preis * gewicht # berechnung des gesamtpreises
    produkte_mit_gesamtpreis.append((name, preis, gewicht, gesamtpreis)) # hinzufügen ALLER Felder in eine neue Liste

# Sortieren nach Gesamtpreis
produkte_mit_gesamtpreis.sort(key= lambda x: x[3]) # wichtig!! lambda x steht für das objekt, x[] spezifiziert den INDEX des Elements im Objekt

# Ausgabe der sortierten Liste mit Gesamtpreisen
print(produkte_mit_gesamtpreis) # simplified, hier nochmal gerne in die Lösung gucken damit das der erwarteten Ausgabe entspricht
```

## Lösung zu Aufgabe 2

```python
# Aufgabenliste mit Prioritäten
aufgaben = [("Einkaufen", 1),("Hausaufgaben machen", 2),("Sport treiben", 3)]

# Indexieren und Prioritätstext zuweisen
for index, (aufgabe, priorität) in enumerate(aufgaben, start=1):
    match priorität:
        case 1:
            priorität_text = "Hoch"
        case 2:
            priorität_text = "Mittel"
        case 3:
            priorität_text = "Niedrig"

    # Ausgabe der Aufgabe mit Priorität
    print(f"{index}. {aufgabe} - Priorität: {priorität_text}")
```

```python
# Musterlösung!! 

aufgaben = [("Einkaufen", 1),("Hausaufgaben machen", 2),("Sport treiben", 3)]

for index, (aufgabe, priorität) in enumerate(aufgaben, start=1):
    if priorität == 1:
        priorität_text = "Hoch"    
    elif priorität == 2:
        priorität_text = "Mittel"
    else:
        priorität_text = "Niedrig"

    print(f"{index}. {aufgabe} - Priorität: {priorität_text}")    
```

```python
# Lösung aus dem Tutorium

# Aufgabenliste mit Prioritäten
aufgaben = [("Einkaufen", 1), ("Hausaufgaben machen", 2), ("Sport treiben", 3)] # liste mit tuples

# Ausgabe der Aufgabenliste MIT Prioritäten
for i, (aufgabe, priorität) in enumerate(aufgaben, start=1): # i = index in liste, (element 1, element 2) pro objekt in unserem 
# enumerate(liste, start=$INDEX AB DEM ANGEFANGEN WIRD ZU ZÄHLEN)

    match priorität: # auch möglich: if-else-schleife
        case 1:
            priorität_text = "Hoch"
        case 2:
            priorität_text = "Mittel"
        case 3:
            priorität_text = "Niedrig"

    print(f"Aufgabe {i}: {aufgabe} - (Priorität: {priorität_text})") # wenn nicht "start=1" oben spezifiziert wurde, dann auch "Aufgabe {i+1}" mgl.
```

## Lösung zu Aufgabe 3

```python
datenbestand = {
    "Apfel": "Obst",
    "Karotte": "Gemüse",
    "Wasser": "Getränk",
    "Banane": "Obst",
    "Tomate": "Gemüse",
    "Club Mate": "Spaßgetränk",
    "Orange": "Obst",
    "Brokkoli": "Gemüse",
    "Tee": "Getränk",
    "Traube": "Obst"
}

# Leeres dict für Zähler
kategorien = {}

# Zählen der Kategorien
for artikel, kategorie in datenbestand.items():
    if kategorie in kategorien:
        kategorien[kategorie] += 1
    else:
        kategorien[kategorie] = 1

# Ausgabe der Zählergebnisse
for kategorie, anzahl in kategorien.items():
    print(f"Kategorie: {kategorie}, Anzahl der Objekte: {anzahl}")
```
