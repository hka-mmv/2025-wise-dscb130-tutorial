# Python – Strings Methods Grundlagen



## Was sind Strings?
- Text in Python
- immer zwischen `" "` oder `' '`
- Beispiele: `"Hallo"`, `'Python'`, `"123"`


```python
Wort = "Hallo"
Wort2 = 'Test'
Satz = "Matthias ist der beste Dozent der Welt"
Zahl = "56"

type(Wort)
```


## Warum Strings?
- Namen speichern
- Texte verarbeiten (z.B. Chat)
- Viele Daten kommen als Text



## Strings sind immutable
man verändert sie nicht direkt – man erzeugt NEUE


## .upper()

Verwandelt ALLES in Großbuchstaben

```python
s = "marc"

print(s)   # "marc"

s = s.upper()
print(s)   # "MARC"
```

## .lower()

Verwandelt ALLES in Kleinbuchstaben

```python
name = "Marc"

print(name.upper())   # MARC

print(name.lower())   # marc
```

## .strip() – Leerzeichen entfernen

Entfernt nur ÄUßERE Leerzeichen!

```python
text = "         Test        "

clean = text.strip()

print(clean) 
```

```python
text2 = "   Hallo Welt   "

text2_stripped = text2.strip()

print(text2_stripped)
```

## .replace() – Text tauschen

Kann Teile des Strings komplett ersetzen .replace(erstes Argument, zweites Argeument)
Erstes Argument = zu ersetzen
Zweites Argument = mit was ersetzt wird

```python
msg = "Hallo Welt"

neu = msg.replace("Welt", "Marc")

print(neu)    # Hallo Marc
```

## .split() – String -> Liste

Teilt den String am Teiler Argument. Die aufgeteilten Einzelteile werden in eine Liste gepackt.

```python
satz = "Ich;mag;Pizza"

teile = satz.split(";")

print(teile)
```

## .join() – Liste -> String

Verbindet Einzelteiele in einer Liste zu einem String. Als erstes mit welchem Stringzeichen sie verbunden werden sollen. Im Argument dann welches Objekt verbundne werden soll

```python
teile = ['Ich', 'mag', 'Pizza']

# mit Leerzeichen verbunden " "
neu = " ".join(teile) # Liste teile soll verbunden werden

print(neu) 
```

## Bonus: f-Strings

Kann veränderbare Objekte drucken. Ist dynamischer und dadurch widerverwendbarer.

```python
# schlecht hard coded (ohne f-string)
print("Ich bin Marc und 16 Jahre alt.")


name = "Marc"
alter = 16

# gut dynamisch gecoded (mit f-string)
print(f"Ich bin {name} und {alter} Jahre alt.")
```
