## Tkinter

Tkinter ist eine Built in Python Library(muss nicht heruntegeladen werden) mit welcher man eine grafische Oberfläche erzeugen kann(GUI = Graphical User Interface).

```python
# Tkinter muss importiert werden. Am besten ganz oben im Prgramm.
import tkinter as tk
# tk ist ein Alias(frei Wählbarer Name) mit dem Funktionen dieser library im Prgramm aufgerufen werdn können.
```

## 1. Fenster aufrufen

```python
# Fenste wird generiert
fenster = tk.Tk()

# fenster bennenen 
fenster.title("Mein erstes GUI")

# Fenster Größe (Pixel X-Achse, Pixel >-Achse)
fenster.geometry("650x450")

# Mainloop hält Fenster am laufen(Alle Fenster Logik muss VOR mainloop passiern)
# Wenn Mainloop beendet -> restlicher Python code wird ausgeführt
fenster.mainloop()

print("Mainloop beendet")
```

## 1.1 Widgets
Widgets sind Bausteine der Oberfläche. Können z.B. Textfelder oder Knöpfe sein etc.

```python
# 1. Label ist eine einfache Textanzeige

# label = tk.Label(parent, text="Mein Text")
label = tk.Label(fenster, text="wichtiges Textfeld")
# muss mit grid extra aufgerufen werden, für späteres Platzierung
label.grid()


# 2. Entry = kann Text entgegen nehmen oder anzeigen
# var = tk.StringVar(Ausgabewert)
# display = tk.Entry(parent, textvariable=var)


var = tk.StringVar(value="Neue Text Anzeige")
display = tk.Entry(fenster, textvariable=var)
display.grid() # für Raster Platzierung


# 3. Button = klicken des Buttons kann mit Funktion verbunden werden
# button = tk.Button(parent, text="Erster Button", command=Variable/Funktion/...)

meine_funktion = 1 + 1

button = tk.Button(fenster, text="Erster Button", command=meine_funktion)
button.grid()  # für Raster Platzierung

```

## 3. Layout Management
Die bisher frei "schwebenden" Widgets(Labels/Entrys/Buttons) müssen noch strukturiert plaziert werden. Ein guter Raster Manager ist die grid() Funktion. Dieser ordnet die Widgets in einem "unsichtbaren" Tabellen Raster an.

```python
# Das Display wird durch die grid() Funktion einem Platz zugewiesen
# grid(Reihem, Spalte, Spalten verbinden, Pixel X-Achse, Pixel Y-Achse)
display.grid(row=10, column=10, columnspan=4, padx=5, pady=5)
# Zeilen und Spalten Anzahl wird durch Platzierung erzegut. Hier existieren jetzt 11 Zeilen und Spalten (0 Index)

# so lassen sich alle widgets platzieren
button.grid(row=6, column=10, columnspan=1, padx=10, pady=2)
```
