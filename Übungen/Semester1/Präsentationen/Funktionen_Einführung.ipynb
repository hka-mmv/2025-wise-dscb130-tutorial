# Funktionen
- eine Funktion ist ein benannter Codeblock, welcher wieder verwendet werden kann
- kann Werte ausgeben
- kann Werte entgegen nehmen, dann verarbeiten und die verarbeitetn Werte ausgeben
- Vorteile: Wiederverwendbarkeit, bessere Struktur, Wartbarkeit
- bekannte (built in)Funktionen: print(), input(), sum(), type()

```python
# OHNE Funktion
print("Hallo Tom")
print("Hallo Felix")
print("Hallo Lena")
```

```python
# MIT Funktion
def begrüßung(name):
    print(f"Hallo {name}")

begrüßung("Peter")
```

## Syntax

- def = define: definiere neue Funktion
- funktionsname: kann man frei wählen, sollte aber die Funktion beschreiben
- Parameterliste: Werte festlegen welche der Benutzer der Funktion einfügen muss damit sie verarbeite werden. (Parameter1, Parameter2, ...)
- Docstring: drei Anführungsstriche vorne und hinten, funktioniert wie ein Kommentar und sollte die Funktion erklären
- Funktionstext: Hier wird programmiert was die Funktion tatsächlich macht, z.B. ihre Parameter verarbeiten
- return: beendet die Funktion! Gibt einen weiterverwendbaren Rückgabewert zurück, z.B die verarbeiteten Parameter. (Wenn kein return = None)

## Funktion aufrufen
Eine Funktion wird erst ausgeführt wenn sie aufgerufen wird.

```python
# oben definierte Funktion wird hier aufgerufen
begrüßung("Paul")
# Paul ist das Argument welches in den Parameter "name" eingefügt wird
```

```python
# Aufruf mit mehreren Argumenten
def addition(a, b):
    """Addiert Argument a und b zusammen"""
    ergebnis = a + b
    return ergebnis

# Aufruf
addition(5, 10)
# Argumente können auch expliziet benannt werden
addition(b=5, a=3)
```

```python
# Parameter kann einen Standardwert als Argument haben, falls keine Argumente übergeben werden
def begrüßung2(name="Gast"):
    print("Hallo, wie geht es dir", name)

begrüßung2() # Standardwert
begrüßung2("Tim") # Standardwert wird überschrieben
```

## Scope
Bestimmt wo Variablen gültig sind.
- Lokaler Scope = innerhalb einer Funktion
- Globaler Scope = außerhalb einer Funktion

```python
name = "Anna"        # globale Variable

def begrüßung3():
    name = "Ben"     # lokale Variable
    print(name)

begrüßung3() # Aufruf der Funktion
print(name) # Ausgabe der Variable
```

## Kennzeichnen von Funktionen
Wie eine Funktion am besten "beschrifete" wird, damit auch andere deinen Code verstehen

### Positional Arguments
Sind die Argumente welche in die Funktion gegeben werden. Müssen die richtige Reihenfolge haben!

```python
def begrüßung4(name, alter):
    print(f"Hallo {name}, du bist {alter} Jahre alt.")

begrüßung4("Anna", 25)   # funktioniert
begrüßung4(25, "Anna")   # falsche Reihenfolge → andere Bedeutung / evtl. Fehler
begrüßung4("Anna")       # Fehler: age fehlt
```

### Default Arguments
Standardwerte eines Parameters wenn KEIN Argument übergeben wird. Default Argumente dürfen immer nur nach Positional Argument kommen!

```python
#                   default argument kommt als letztes
def begrüßung5(name, alter=18):
    print(f"Hallo {name}, du bist {alter} Jahre alt.")

begrüßung5("Anna")          # nutzt default age=18
begrüßung5("Ben", 30)       # überschreibt default
```

### Type Hints
Sind komplett optional, Python bleibt dynamisch, geben nur Hinweise darauf welche Typen eingegeben oder ausgegeben werden sollten.

```python
# Arguemnte sollten int sein
def multiplikation(a: int, b: int) -> int:  # Der Output sollte auch int sein
    return a * b

multiplikation(5, 10)
```
