# Python – Strings Methods Grundlagen Übungen

```python
# 1. .upper() und .lower() (Groß- und Kleinschreibung)
```

```python
# Aufgabe 1.1: Schreibe das Wort wort in Großbuchstaben.
```

```python
wort = "klein"

print(wort.upper())

print(wort)
```

```python
# Aufgabe 1.2: Schreibe den Namen name in Kleinbuchstaben.
```

```python
name = "ANNA"


print(name.lower())


print(name)
```

```python
# 2. .strip() (Leerzeichen entfernen)
```

```python
# Aufgabe 2.1: Entferne die Leerzeichen am Anfang und Ende von zahl.
```

```python
zahl = " 42 "


print(zahl.strip())


print(zahl)
```

```python
# Aufgabe 2.2: Entferne nur das rechte (nachfolgende) Leerzeichen von ort.
```

```python
ort = "Wien. "


print(ort.rstrip())


print(ort)
```

```python
# 3. .replace() (Text tauschen)
```

```python
# Aufgabe 3.1: Ersetze das Wort "Herr" durch "Frau".
```

```python
text = "Guten Tag Herr Schmidt"


print(text.replace("Herr", "Frau"))


print(text)
```

```python
# Aufgabe 3.2: Ersetze alle Bindestriche (-) durch Kommas (,).
```

```python
farbe = "rot-blau-gruen"


print(farbe.replace("-", ","))


print(farbe)
```

```python
# 4. .split() (String $\rightarrow$ Liste)
```

```python
# Aufgabe 4.1: Teile den String in eine Liste, getrennt durch das Komma (,).
```

```python
fische = "Hai,Wal,Delfin"


print(fische.split(","))


print(fische)
```

```python
# Aufgabe 4.2: Teile den Pfad in eine Liste, getrennt durch den Schrägstrich (/).
```

```python
pfad = "ordner/datei.txt"


print(pfad.split("/"))


print(pfad)
```

```python
# 5. .join() (Liste $\rightarrow$ String)
```

```python
# Aufgabe 5.1: Verbinde die Namen in der Liste mit einem Leerzeichen (" ") dazwischen.
```

```python
namen = ["Max", "Mia", "Leo"]


print(" ".join(namen))


print(namen)
```

```python
# Aufgabe 5.2: Verbinde die Ziffern zu einem String ohne Trennzeichen (verwende "").
```

```python
ziffern = ["1", "9", "2", "3"]


print("".join(ziffern))


print(ziffern)
```

```python
# 6. f-Strings (Formatierung)
```

```python
# Aufgabe 6.1: Erstelle einen f-String, der die Ausgabe "Ich habe 10 Äpfel gekauft." erzeugt.
```

```python
zahl = 10


text = "Äpfel"


print(f"Ich habe {zahl} {text} gekauft.")
```

```python
# Aufgabe 6.2: Erstelle einen f-String, der den preis in die Ausgabe einfügt.
```

```python
preis = 19.99


print(f"Das Produkt kostet {preis} Euro.")
```
